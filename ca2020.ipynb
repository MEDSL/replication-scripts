```python
def fixPrecinct(x, a, b, c, d):
    return x + '_' + str(d).zfill(3) + str(b).zfill(3) + str(c).zfill(3) + str(a).zfill(3)
df['precinct'] = df.apply(lambda x: fixPrecinct(x['precinct'], x['ADDIST'], x['CDDIST'], x['SDDIST'], x['BEDIST']), axis=1)
```

```python
def office_codebook(x, y, z, a):
    if 'ASS' in x:
        office_codebook = x[0:3] + str(y).zfill(2) + x[3:8]
    elif 'CNG' in x:
        office_codebook = x[0:3] + str(z).zfill(2) + x[3:8]
    elif 'SEN' in x:
        office_codebook = x[0:3] + str(a).zfill(2) + x[3:8]
    else:
        office_codebook = x
    return office_codebook
df['office_codebook'] = df.apply(lambda x: office_codebook(x['office'], x['ADDIST'], x['CDDIST'], x['SDDIST']), axis=1)
# df
```

```python
def getParty(x):
    if 'DEM' in x:
        return 'DEMOCRAT'
    if 'REP' in x: 
        return 'REPUBLICAN'
    if 'GRN' in x:
        return 'GREEN'
    if 'IND' in x:
        return 'INDEPENDENT'
    if 'LIB' in x:
        return 'LIBERTARIAN'
    if 'AIP' in x:
        return 'ALLIANCE'
    if 'PAF' in x:
        return 'SOCIALISM AND LIBERATION'
    else:
        return ''
    
df['party_detailed'] = df['office'].apply(getParty)
df['party_simplified'] = df['party_detailed'].replace({'GREEN': 'OTHER', 'INDEPENDENT': 'OTHER', 
                                                       'SOCIALISM AND LIBERATION' : 'OTHER', 'ALLIANCE' : 'OTHER'})
```

```python
def cleanCandidate(x, y):
    x = x.replace('*', '')
    x = x.replace('.', '')
    x = x.strip()
    if '_Y' in y:
        return 'YES'
    if '_N' in y:
        return 'NO'
    return x
df['candidate'] = df.apply(lambda x: cleanCandidate(x['candidate'], x['office']), axis=1)
df
```

```python
def cleanOffice(x):
    if 'ASS' in x:
        return 'STATE HOUSE'
    if 'CNG' in x:
        return 'US HOUSE'
    if 'SEN' in x:
        return 'STATE SENATE'
    if 'PRS' in x:
        return 'US PRESIDENT'
    if 'PR_' in x:
        x = x[0:5]
        x = x.replace('_', ' ')
        x = x.replace('PR', 'PROPOSITION')
        return x
    else:
        return x
df['office'] = df['office'].apply(cleanOffice)
```

```python
def getDistrict(a, b, c, d):
    if 'STATE HOUSE' in a:
        return str(b).zfill(3)
    if 'US HOUSE' in a:
        return str(c).zfill(3)
    if 'STATE SENATE' in a:
        return str(d).zfill(3)
    if any(s in a for s in['PROPOSITION', 'PRESIDENT']):
        return 'STATEWIDE'
    else:
        return ''
df['district'] = df.apply(lambda x: getDistrict(x['office'], x['ADDIST'], x['CDDIST'], x['SDDIST']), axis=1)
```

```python
def getDataverse(x):
    if any(s in x for s in ['STATE', 'PROPOSITION']):
        return 'STATE'
    if 'PRESIDENT' in x:
        return 'PRESIDENT'
    if 'US HOUSE' in x:
        return 'HOUSE'
    else:
        return ''
df['dataverse'] = df['office'].apply(getDataverse)
```

```python
def getMagnitude(x):
    if 'REGISTERED VOTERS' in x:
        return 0
    else:
        return 1
df['magnitude'] = df['office'].apply(getMagnitude)
```
