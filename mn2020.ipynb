```python
def getOfficeName(x):
    if 'U.S. Representative' in x: 
        return 'US HOUSE'
    elif 'State Senator' in x: 
        return 'STATE SENATE'
    elif 'State Representative' in x: 
        return 'STATE HOUSE'
    else:
        return x
```

```python
candCodes['office'] = ''
candCodes['office'] = candCodes['officeName'].apply(getOfficeName)
candCodes['office'] = candCodes['office'].str.strip()
```

```python
def getMergeCodes(x):
    if ')' not in x: 
        newX = x[-3:]
        if 't' in newX or 's' in newX: 
            newX = newX[1:]
    else: 
        index = x.index('(')
        newX = x[index-2:index-1]
    return newX
```

```python
candCodes['mergeCode'] = ''
candCodes['mergeCode'] = candCodes['officeName'].apply(getMergeCodes)
candCodes['mergeCode'] = candCodes['mergeCode'].str.strip()
candCodes['PartyAbbv'] = candCodes['PartyAbbv'].str.strip()
```
